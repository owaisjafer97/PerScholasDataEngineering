```python
#importing all required libraries and classes
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read Data from the woocommerce-product-export.csv file using the python library.
df = pd.read_csv('woocommerce-product-export.csv')
```

```python
#Print  the "total_profit," and “month_number” columns only.
print(df[['total_profit','month_number']])
```

```python
#Read the total profit of all months and show it using the Bar plot. 
#The generated Bar plot must include the following properties: 
#X label name = Month number.
#Y label name = Total profit.
#Title = Company profit per month.
X = df['month_number'] #X Axis
Y = df['total_profit'] #Y Axis

ax = plt.axes()
plt.bar(X,Y)
plt.xlabel("Month number")
plt.ylabel("Total profit")
plt.title("Company profit per month")

#to add grids to the graph
ax.yaxis.grid(True, linestyle = '--')
ax.xaxis.grid(True, linestyle = '--')

```

```python
#Read the total profit of all of the months, and show the line plot with the following style properties:
#The generated line plot must include the following Style properties: 
#The line style should be dotted and the line color should be red.
#Show the legend at the lower-right location.
#X label name = Month number
#Y label name = Total profit.
#Title = Company profit per month.
#Add a circle marker; the line marker color = black. 
#The line width should be 3.
Y = df['total_profit']
X = df['month_number']
plt.plot(X,Y,linestyle = "--", linewidth = '3', color = 'red', marker = 'o', markerfacecolor = 'black')
plt.xticks(np.arange(1,13,1))
plt.xlabel("Month number")
plt.ylabel("Total profit")
plt.title("Company profit per month")
plt.legend(["Profit data of last year"],loc = 'lower right')
```

```python
#Print all of the product sales data and show it using a multi-line plot.
#Display the number of units sold per month for each product using multi-line plots. (i.e., show a separate Plotline for each product).
#The graph should look like the screenshot below:
#GRAPH BELOW SHOULD NOT HAVE TOOTHPASTE MULTIPLE TIMES IN LEGEND. DISREGARD THAT IN THE LEGEND, ONE LINE FOR EACH PRODUCT.
facecream = df['facecream']	
facewash = df['facewash']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']
X = df['month_number']

plt.plot(X,facecream, linewidth = '3', color = 'red', marker = 'o', label = "Face cream Sales Data")
plt.plot(X,facewash, linewidth = '3', color = 'blue', marker = 'o', label = "Face wash Sales Data") #set alpha (transparency) = 0 if you want to see moisturizer sales
plt.plot(X,toothpaste, linewidth = '3', color = 'green', marker = 'o', label = "Tooth paste Sales Data")
plt.plot(X,bathingsoap, linewidth = '3', color = 'yellow', marker = 'o', label = "Bathing soap Sales Data")
plt.plot(X,shampoo, linewidth = '3', color = 'orange', marker = 'o', label = "Shampoo Sales Data")
plt.plot(X,moisturizer, linewidth = '3', color = 'black', marker = 'o', label = "Moisturizer Sales Data") #set alpha (transparency) = 0 if you want to see face wash sales
plt.xticks(np.arange(1,13,1))
plt.xlabel("Month number")
plt.ylabel("Sales units in number")
plt.title("Sales data")
plt.legend()

#Face wash and Moisturizer have the exact same numbers when it comes to sales hence there is an overlap, hence blue line cannot be seen. If you want to see facewash sales set alpha = 0 for moisturizer plot and if you want to see moisturizer sales set facewash plot alpha = 0
```

```python
#Read “bathingsoap” sales data for each month and show it using a scatter plot.
#Add a grid to the plot. The grid line style should be “–“.

X = df['month_number']
Y = df['bathingsoap']
ax = plt.axes()
plt.scatter(X,Y)
plt.xticks(np.arange(1,13,1))
plt.xlabel("Month number")
plt.ylabel("Sales units sold")
plt.title("Bathing soap Sales data")
ax.yaxis.grid(True, linestyle = '--')
ax.xaxis.grid(True, linestyle = '--')
```
